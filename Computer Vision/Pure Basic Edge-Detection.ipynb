# Edge Detection Using verticle and horizontal edge detection.

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
k = np.array([[0, -1, 0],
             [-1, 4, -1],
             [0, -1, 0]])

a = np.array([[ -1, -2, -1], 
              [ 0, 0, 0], 
              [ 1, 2, 1]])


b = a.T

print(b)
```

```python
detect_v = filtered_image = cv2.filter2D(gray, -1, a)

plt.imshow(filtered_image, cmap='gray')
```

```python
retval, binary_image_v = cv2.threshold(detect_v, 100, 255, cv2.THRESH_BINARY)

plt.imshow(binary_image_v, cmap='gray')
```

```python
detect_h = filtered_image = cv2.filter2D(gray, -1, b)

plt.imshow(filtered_image, cmap='gray')
```

```python
retval, binary_image_h = cv2.threshold(detect_h, 100, 255, cv2.THRESH_BINARY)

plt.imshow(binary_image_h, cmap='gray')
```

```python
total_det = (binary_image_v) + (binary_image_h)

plt.imshow(total_det, cmap='gray')
```

```python
total_det = np.multiply(binary_image_v,binary_image_v) + np.multiply(binary_image_h,binary_image_h)
print(type(total_det))
plt.imshow(total_det, cmap='gray')
```

```python
total_det = np.multiply(total_det,total_det)
plt.imshow(total_det, cmap='gray')
```

# Final Edge Detection function.

```python
def edge_detect(image):
  gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
  
  k = np.array([[0, -1, 0],
              [-1, 4, -1],
              [0, -1, 0]])

  a = np.array([[ -1, -2, -1], 
                [ 0, 0, 0], 
                [ 1, 2, 1]])

  b = a.T

  detect_v = filtered_image = cv2.filter2D(gray, -1, a)
  retval, binary_image_v = cv2.threshold(detect_v, 100, 255, cv2.THRESH_BINARY)
  detect_h = filtered_image = cv2.filter2D(gray, -1, b)
  retval, binary_image_h = cv2.threshold(detect_h, 100, 255, cv2.THRESH_BINARY)
  total_det = (binary_image_v) + (binary_image_h)

  plt.imshow(total_det, cmap='gray')
```
