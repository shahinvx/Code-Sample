# Import library.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from imageai.Detection import ObjectDetection

from PIL import Image
import glob
import numpy as np
import os

from kora.drive import upload_public
from IPython.display import HTML
```

# Vehicle detection v2.

## Import Our custom Object detection model.

```python
detector = ObjectDetection()
model_path = "model/sn_custom_v2.h5"
detector.setModelTypeAsYOLOv3()
detector.setModelPath(model_path)
```

## Detect Vehicle from image.

```python
input_path = "Data/frame_1.jpg"
output_path = "out_img/detect_img_2.jpg"
detector.loadModel()
detection = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path, minimum_percentage_probability=65)
print(detection)
```

## Show the detected image.

```python
img = mpimg.imread('out_img/detect_img_2.jpg')
plt.figure(figsize=(12,8))
plt.imshow(img)
```

# Processing this video for detecting and finding average image frame.

```python
url = upload_public('Data/video.mp4')

HTML(f"""<video src={url} width=500 controls/>""")
```

## Random frame show.

```python
img = mpimg.imread('raw_img/7.jpg')
plt.figure(figsize=(12,8))
plt.imshow(img)
```

## Create average frame from whole images of the video.

```python
imgList = glob.glob('raw_img/*.jpg')
first = True

for img in imgList:
    temp = np.asarray(Image.open(img))
    temp = temp.astype('uint32')
    if first:
        sumImage = temp
        first = False
    else:
        sumImage = sumImage + temp

avgArray = sumImage/len(imgList)
avg_image = avgArray.astype('uint8')
```

## Show the final average image.

```python
plt.figure(figsize=(12,8))
plt.imshow(avg_image)
```
